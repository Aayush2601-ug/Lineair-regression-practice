```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd

data = pd.read_csv('headbrain.csv')

# Extract and reshape the data
x = data['Head Size(cm^3)'].values.reshape(-1, 1)
y = data['Brain Weight(grams)'].values.reshape(-1, 1)  # Reshape y to match the dimensions of x

model = LinearRegression()
model.fit(x, y)


predictions = model.predict(x)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual')  # Use the reshaped x and y here
plt.plot(x, predictions, color='red', linewidth=2, label='Predicted')
plt.title('Actual vs Predicted Brain Weights')
plt.xlabel('Head Size (cm^3)')
plt.ylabel('Brain Weight (grams)')
plt.legend()
plt.grid(True)
plt.show()

```
